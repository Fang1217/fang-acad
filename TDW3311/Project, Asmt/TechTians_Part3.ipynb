## Part 3- Data Visualisation (5 Marks)
Select any TWO numerical variables from `Dataset1`. Use Pandas to create a histogram to visualise the distribution of each of these variables.\
Briefly describe each histogram.

```python
import pandas as pd
import matplotlib.pyplot as plt

df3 = pd.read_csv('Dataset1.csv', header = 0)
```

### 1. Histogram of Latitude

```python
# Histogram of Latitude

plt.hist(df3['latitude'], color = 'skyblue', ec = 'black', bins = 50)
plt.title('Histogram of Latitude')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.show()
```

#### <u>Description of histogram of Latitude</u>:
Since the histogram is not in bell-shape, so it is a non-normal histogram.\
In addition, it is a long tailed histogram as the tails approach zero very slow.

### 2. Histogram of Price

```python
# Histogram of Price

plt.hist(df3['price'], color = 'green' , ec = 'black', bins = 50)
plt.title('Histogram of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')

plt.show()
```

```python
# Edited histogram of price, given the graph is leaning towards the left while having other empty bins,
# specify the bins to be between 0 to 1200. with 50 steps.

bins = []
for i in range(50):
    bins.append(i*25)

plt.hist(df3['price'], color = 'green' , ec = 'black', bins = bins)
plt.title('Histogram of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')

plt.show()
```

#### <u>Description of histogram of Price</u>:
This is a skewed right histogram as most of the data falls to the right of the graph's peak.\
 This showed that the the majority of prices are concentrated at the low end.
